```python
import pandas as pd
filename = r'C:\\sqlite\\mysql\\code\\Code11\\file2\\dust.xlsx'
df = pd.read_excel(filename, engine='openpyxl')
df
```

```python
df.rename(columns={'날짜':'date', '아황산가스':'SO2','일산화탄소':'CO', '오존':'O3','이산화질소':'NO2'},inplace=True) # inplace 원본수정허용
df
```

```python
df['date'] = df['date'].str[:10]
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['date']
```

```python
df['year'] = df['date'].dt.year #연도
df['month'] = df['date'].dt.month #월
df['day'] = df['date'].dt.day #일
```

```python
df = df[['date', 'year', 'month', 'day', 'SO2', 'CO', 'O3', 'NO2', 'PM10', 'PM2.5']]
```

```python
df.isnull().sum()
```

```python
# 시계열분석

df = df.fillna(method = 'pad')
df.head()
```

```python
# 이전값이 없는경우 혼자NaN은 20으로 채워줌
df.fillna(20, inplace=True)
df.head()
```
