```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_excel('C:\\sqlite\\mysql\\code\\07_1\\file\\health_screenings_2020_1000ea.xlsx')
```

```python
data6 = data.loc[:,['gender', 'height', 'weight', 'waist', 'drinking', 'smoking']]
data6.loc[data6['gender']==1,['gender']]='Male'
data6.loc[data6['gender']==2,['gender']]='Female'
data6.loc[data6['drinking']==0,['drinking']]='Non-drinking' #data6 drinking 0 Non-drinking   1  Drinking  .
data6.loc[data6['drinking']==1,['drinking']]='Drinking'#data6 drinking 0 Non-drinking   1  Drinking  .
data6.loc[(data6['smoking']==1) | (data6['smoking']==2),['smoking']]='Non-smoking'
data6.loc[data6['smoking']==3,['smoking']]='Smoking'
```

```python
drinking = data6.groupby(['gender', 'drinking'])['drinking'].count() # gender, drinking  drinking count
# print(drinking)
smoking = data6.groupby(['gender', 'smoking'])['smoking'].count() #  gender,smoking  smoking count
# drinking = drinking.to_frame(name='gender')
drinking = drinking.to_frame(name='count') #      count
# print(drinking)
smoking = smoking.to_frame(name='count') #      count
drinking = drinking.reset_index() #  
print(drinking)
smoking = smoking.reset_index()
```

```python
fig = plt.figure(figsize=(17, 6))
fig.suptitle('2020 Health', fontweight = 'bold')
index = np.arange(4)

fig.add_subplot(1,2,1)

plt.bar(index, drinking['count'])
plt.title('drinking')
plt.ylabel('Count')
plt.xticks(index, ['Non-drinking(Female)', 'Drinking(Female)', 'Non-drinking(Male)', 'Drinking(Male)'])

fig.add_subplot(1,2,2)

plt.bar(index, smoking['count'])
plt.title('Smoking Type')
plt.ylabel('Count')
plt.xticks(index, ['Non-drinking(Female)', 'Drinking(Female)', 'Non-drinking(Male)', 'Drinking(Male)'])

plt.show()
```

```python
fig = plt.figure(figsize=(17, 6))

area1 = fig.add_subplot(1,2,1)
area2 = fig.add_subplot(1,2,2)

ax1 = sns.barplot(data=drinking, x='gender', y='count', hue='drinking', ax=area1)
ax2 = sns.barplot(data=smoking, x='gender', y='count', hue='smoking', ax=area2)

fig.suptitle('2020 Health', fontweight='bold')
area1.set_title('drinking type')
area2.set_title('smoking type')

plt.show()
```
