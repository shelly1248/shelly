```python
import pandas as pd

df = pd.read_excel('C:\\sqlite\\mysql\\code\\AI\\Mid_project\\data\\all.xlsx')
df.drop('Unnamed: 0', axis=1, inplace=True)
df_경상 = df[df['피해운전자 상해정도'] == '경상']
df_중상 = df[df['피해운전자 상해정도'] == '중상']
df_부상신고 = df[df['피해운전자 상해정도'] == '부상신고']
df_사망 = df[df['피해운전자 상해정도'] == '사망']

total_경상 = df_경상.groupby('시/구').size()
total_중상 = df_중상.groupby('시/구').size()
total_부상신고 = df_부상신고.groupby('시/구').size()
total_사망 = df_사망.groupby('시/구').size()

total = total_경상.add(total_중상, fill_value=0)
total = total.add(total_부상신고, fill_value=0)
total = total.add(total_사망, fill_value=0)
total.reset_index()
df['total'] = total[0]
df_all = pd.read_excel('C:\\sqlite\\mysql\\code\\AI\\total좌표.xlsx')
columns = ["주소", "경도", "위도"]
df_all.columns = columns
df['경도'] = df_all['경도']
df['위도'] = df_all['위도']
###################################################################
import geopandas as gpd
import folium
from folium.plugins import HeatMap
from shapely.geometry import Point
from branca.colormap import LinearColormap
from folium.plugins import MarkerCluster

df['시/구total'] = df.groupby('시/구')['피해운전자 상해정도'].transform('count')

geometry = [Point(lon, lat) for lat, lon in zip(df['위도'], df['경도'])]
gdf_coordinates = gpd.GeoDataFrame(df, geometry=geometry, crs='EPSG:4326')

gdf_coordinates = gdf_coordinates[~gdf_coordinates['geometry'].is_empty]

total = gdf_coordinates.groupby('geometry')['피해운전자 상해정도'].count().reset_index(name='total')
gdf_coordinates = gdf_coordinates.merge(total, on='geometry')

mymap = folium.Map(location=(37,127), zoom_start=12, control_scale=True)

marker_cluster = MarkerCluster().add_to(mymap)

for index, row in gdf_coordinates.iterrows():
    if not row['geometry'].is_empty:
        latitude = row['geometry'].y
        longitude = row['geometry'].x
        total_count = row['시/구total']
        popup_text = f'Total Counts: {total_count}'
        
        icon = folium.Icon(color='red', icon='info-sign')
        
        folium.Marker(location=[latitude, longitude], popup=popup_text, icon=icon).add_to(marker_cluster)

# Save the map
mymap.save('C:\\sqlite\\mysql\\code\\AI\\Mid_project\\data\\file\\cartoon_map_with_injury_counts.html')
```
