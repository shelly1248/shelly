```python
import geopandas as gpd
from shapely.geometry import Polygon
import json
import re


path = r'C:\\sqlite\\mysql\\code\\AI\\Mid_project\\data\\file\\'

with open(path + '전국어린이보호구역표준데이터.json', 'r', encoding='UTF8') as j:
    contents = json.load(j)

records = contents['records']

polygon_list = []

for record in records:
    lon = float(record['경도'])
    lat = float(record['위도'])
    coords = [(lon-0.0001, lat-0.0001), (lon+0.0001, lat-0.0001),
              (lon+0.0001, lat+0.0001), (lon-0.0001, lat+0.0001)]
    polygon = Polygon(coords)
    polygon_list.append(polygon)

geo_df = gpd.GeoDataFrame(records, geometry=polygon_list)

def get_province(name):
    match = re.match(r'^(서울특별시|부산광역시|대구광역시|인천광역시|광주광역시|대전광역시|울산광역시|세종특별자치시|경기도|강원도|충청북도|충청남도|전라북도|전라남도|경상북도|경상남도|제주특별자치도)', name)
    if match:
        return match.group(1)
    return None

# geo_df['도'] = geo_df['제공기관명'].str.split(' ').str[0]
geo_df['도'] = geo_df['제공기관명'].apply(get_province)

geo_df = geo_df[(geo_df['도'] == '경기도') | (geo_df['도'] == '서울특별시')]

geo_df.to_file('C:\\sqlite\\mysql\\code\\AI\\gyeonggi_protect_polygons.geojson', driver='GeoJSON')
```

```python
import pandas as pd
import geopandas as gpd
import folium
from shapely.geometry import Point
from folium.plugins import MarkerCluster

df = pd.read_excel('C:\\sqlite\\mysql\\code\\AI\\Mid_project\\data\\file\\all_test - 복사본.xlsx')

grouped_df = df.groupby(['시군구', '피해운전자 상해정도']).size().reset_index(name='상해정도 갯수')

grouped_df['피해운전자 상해정도'] = grouped_df['피해운전자 상해정도'].astype(str)
grouped_df['상해정도 갯수'] = grouped_df['상해정도 갯수'].astype(str)
grouped_df['상해정도 정보'] = grouped_df[['피해운전자 상해정도', '상해정도 갯수']].apply(lambda x: ': '.join(x), axis=1)

grouped_df_agg = grouped_df.groupby('시군구')['상해정도 정보'].apply(lambda x: '<br>'.join(x)).reset_index()

df_all = pd.read_excel('C:\\sqlite\\mysql\\code\\AI\\total좌표.xlsx')
columns = ["시군구", "경도", "위도"]
df_all.columns = columns
first_all = df_all.groupby('시군구').first().reset_index()

first_all_merged = first_all.merge(grouped_df_agg, on='시군구', how='left')

geometry = [Point(lon, lat) for lat, lon in zip(first_all_merged['위도'], first_all_merged['경도'])]
gdf_coordinates = gpd.GeoDataFrame(first_all_merged, geometry=geometry, crs='EPSG:4326')
gdf_coordinates = gdf_coordinates[~gdf_coordinates['geometry'].is_empty]

mymap = folium.Map(location=(37.53,127), zoom_start=10, control_scale=True)
marker_cluster = MarkerCluster().add_to(mymap)

for index, row in gdf_coordinates.iterrows():
    if not row['geometry'].is_empty:
        latitude = row['geometry'].y
        longitude = row['geometry'].x
        icon = folium.Icon(color='red', icon='info-sign')
        popup_text = row['상해정도 정보']
        folium.Marker(location=[latitude, longitude], icon=icon, popup=folium.Popup(popup_text, max_width=250)).add_to(marker_cluster)

first_all_merged['상해정도 갯수'] = grouped_df['상해정도 갯수']


first_all_merged['상해정도 갯수'] = first_all_merged['상해정도 갯수'].astype(int)

for index, row in first_all_merged.iterrows():
    latitude = row['위도']
    longitude = row['경도']
    injury_count = row['상해정도 갯수'] 

    radius = 50 

    folium.CircleMarker(
        location=[latitude, longitude],
        radius=radius,
        color='yellow',
        fill=True,
        fill_color='yellow',
        fill_opacity=0.1
    ).add_to(mymap)

gdf_protect = gpd.read_file('C:\\sqlite\\mysql\\code\\AI\\gyeonggi_protect_polygons.geojson')
folium.GeoJson(
    gdf_protect,
    name='protection zones',
    style_function=lambda x: {'fillColor': '#FF0000', 'fillOpacity': 0.5}
).add_to(mymap)



mymap.save('C:\\sqlite\\mysql\\code\\AI\\Mid_project\\data\\file\\cartoon_map_with_injury_counts_test.html')
```
