```python
import pandas as pd

filename = r'C:\\sqlite\\mysql\\code\\AI\\Mid_project\\num-data\\경기-어린이사고20-22.csv'

df = pd.read_csv(filename)
df
```

```python
df = df[df['피해운전자 연령'] <= '12세']
```

```python
df
```

```python
filename_2 = 'C:\\sqlite\\mysql\\code\\AI\\Mid_project\\num-data\\경기-어린이보호구역사고20-22.csv'

df_2 = pd.read_csv(filename_2)
df_2
```

```python
df_2 = df_2[df_2['피해운전자 연령'] <= '12세']
```

```python
df_2
```

```python
merge_df = pd.merge(df, df_2, how='outer', indicator=True)\
    .query("_merge=='left_only'")\
    .drop(columns=['_merge'])
```

```python
merge_df.to_excel('C:\\sqlite\\mysql\\code\\AI\\Mid_project\\num-data\\보호구역외사고.xlsx')
```

```python
merge_df = merge_df.astype({'사고일시':'string'})
```

```python
merge_df['사고일시'] = merge_df['사고일시'].replace({'년':'', '월':'', '일':'', '시':''}, regex=True)
```

```python
merge_df_split = merge_df['사고일시'].str.split(' ', expand=True)
```

```python
merge_df['발생시간'] = merge_df_split[3]
```

```python
merge_df['사고일시'] = merge_df_split[0]+merge_df_split[1]+merge_df_split[2]
```

```python
merge_df['발생시간'].astype({'발생시간':'int64'})
```

```python
merge_df['사고일시']
```

```python
merge_df['발생년월일시'] = pd.to_datetime(merge_df['사고일시'], format='%Y%m%d')
```

```python
merge_df_set = merge_df[['사망자수', '중상자수','경상자수' ,'사고일시']]
```

```python
merge_df_set['사고일시'] = pd.to_datetime(merge_df_set['사고일시'], format='%Y%m%d')
```

```python
merge_df_set['년월'] = merge_df_set['사고일시'].dt.to_period('M')
```

```python
columns_to_sum = ['사망자수', '중상자수', '경상자수']

# '년월' 컬럼으로 그룹화하고 지정한 컬럼의 합계 계산
monthly_sum = merge_df_set.groupby('년월')[columns_to_sum].sum().reset_index()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(17, 7))
sns.lineplot(data=monthly_sum, palette='colorblind', linewidth=2)
plt.rc('font', family='Malgun Gothic')

plt.title('보호구역외사고 20-22', fontsize=20)
plt.xlabel('날짜', fontsize=15)
plt.xticks(rotation = 30)
plt.ylabel('피해자', fontsize=15)
```
