```python
# 제목이 'Limitless' 인 모든 컨텐츠를 찾아보세요
df[df['title']=='Limitless']
```

```python
# 'Robert Rodriguez' 감독이 제작하고 유형이 'Movie'인 모든 컨텐츠를 찾아보세요.
df[(df['director']=='Robert Rodriguez') & (df['type']=='Movie')]
```

```python
# 등록된 날짜가 '2019-07-31'이거나 감독이 'Robert Altman'인 모든 컨텐츠를 찾아보세요.
df[(df['date_added']=='31-Jul-19') | (df['director']=='Robert Altman')]
```

```python
# 감독이 'Orson Welles', 'Aditya Kripalani' 또는'Sam Raimi' 인 모든 컨텐츠를 찾아보세요.
df[(df['director']=='Orson Welles') | (df['director']=='Aditya Kripalani') | (df['director']=='Sam Raimi')]
```
