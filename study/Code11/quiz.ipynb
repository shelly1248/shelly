```python
# 실제 중복데이터 확인
import pandas as pd
infilename = r'C:\\sqlite\\mysql\\code\\Code11\\1023_file\\iris.csv'
df = pd.read_csv(infilename, encoding='cp949')
df[df.duplicated(keep=False)]
```

```python
df2 = df.groupby('Species').mean()
```

```python
# 꽃받침 길이의 시각화
```

```python
# 붓꽃 종류별 꽃 받침 길이의 평균에 대한 시각화
import matplotlib.pyplot as plt
df2.SepalLength[:].plot(kind='bar', rot=0)
plt.title('품종별 그래프')
plt.xlabel('data')
plt.ylabel('꽃받침 길이의 평균값')
plt.show()
```
