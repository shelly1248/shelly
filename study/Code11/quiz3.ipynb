```python
import csv
f = open('C:\\sqlite\\mysql\\code\\Code11\\file2\\gender.csv')
data = csv.reader(f)

m = []
f = []

name = input('     : ')
for row in data :
    if name in row[0] :
        for i in row[3:104] :
            m.append(-int(i))
        for i in row[106:] :
            f.append(int(i))
        break

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize = (10,5), dpi=300)
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name + '     ')
plt.barh(range(101), m, label = '')
plt.barh(range(101), f, label = '')
plt.legend()
plt.show()
```

```python
import csv
f = open('C:\\sqlite\\mysql\\code\\Code11\\file2\\gender.csv')
data = csv.reader(f)

size = []

name = input('     : ')
for row in data :
    if name in row[0] :
        m = 0
        f = 0
        for i in range(101) :
            m += int(row[i+3])
            f += int(row[i+106])
        break
size.append(m)
size.append(f)

import matplotlib.pyplot as plt
plt.rc('font', family ='Malgun Gothic')
color = ['crimson', 'darkcyan']
plt.axis('equal')
plt.pie(size, labels = ['',''], autopct ='%.1f%%', colors = color, startangle =90)
plt.title(name + '    ')
plt.show()
```
