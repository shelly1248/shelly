```python
import pandas as pd
filename = r'C:\\sqlite\\mysql\\code\\수업\\ml\\data\\owid-covid-data.csv'
all_tables = pd.read_csv(filename, parse_dates=['date'])
all_tables
```

```python
korea = all_tables[all_tables.location == 'South Korea']
korea
```

```python
from prophet  import  Prophet

k_covid = pd.DataFrame()
k_covid = korea[['date','total_cases']]
k_covid.rename(columns={'date': 'ds', 'total_cases':'y'}, inplace=True)
m = Prophet()
m.fit(k_covid)
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = m.plot(forecast)
```

```python
fig2 = m.plot_components(forecast)
```

```python
from prophet.plot import plot_plotly, plot_components_plotly
import plotly.graph_objs as go


plot_plotly(m, forecast)
```
